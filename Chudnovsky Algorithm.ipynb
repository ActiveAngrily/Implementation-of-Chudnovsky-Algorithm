<h3>Importing Modules</h3>

```python
import math
from decimal import *
import matplotlib.pyplot as plt
import numpy as np
```

<h3>Defining Functions and Variables</h3>

```python
#reference_pi values taken from piday.org
getcontext().prec = 100000
reference_pi = Decimal('3.')
```

```python
def get_normal(value,maximum, minimum):
    delta = maximum - minimum
    return (value - minimum)/ delta
```

```python
def get_factorial(n):
    out = 1
    for i in range(1,n+1):
        out *= i
    return int(out)
        
```

```python
def get_coefficients(n,precision):
    getcontext().prec = precision
    a = 0
    b = 0
    for k in range(n+1):        
        t1 = (-1)**k
        t2 = get_factorial(6*k)
        t3 = get_factorial(3*k)
        t4 = (get_factorial(k))**3
        t5 = 640320**(3*k)
        
        akn = t1 * t2
        bkn = Decimal(t1 * t2 * k)
        
        abkd = t3 * t4 * t5
        ak = Decimal(akn) / Decimal(abkd)
        bk = Decimal(bkn) / Decimal(abkd)
        a += Decimal(ak)
        b += Decimal(bk)
    return a,b,precision
```

```python
def chudnovsky_algorithm(a,b,precision):
    getcontext().prec = precision
    num = Decimal(426880 * Decimal(10005).sqrt())
    den = Decimal((13591409*a) + (545140134*b))
    pi = Decimal(num / den)
    error = (reference_pi - pi)/reference_pi
    accuracy = 1 - abs(error)
    return pi, Decimal(accuracy * 100 )
```

<h3>Approximation</h3>

```python
iterations = 100
coeffa,coeffb,precision = get_coefficients(iterations,3000)

pi, accuracy = chudnovsky_algorithm(coeffa,coeffb,precision)
accuracy_log = str(accuracy).count("9") / precision
print("precision - '{}' \niterations - '{}'\naccuracy(logarithmic scale) - '{}' \npi - '{}' \n".format(precision, iterations,accuracy_log, pi))
```

<h3>Plotting</h3>

```python
def get_accuracy_list(n):
    acc_list = []
    for i in range(1,n):
        a_i,b_i,prec_i = get_coefficients(i, 10000)
        pi_i, acc_i = chudnovsky_algorithm(a_i,b_i,prec_i)
        acc_list.append(acc_i)
    
    z = []
    
    for i in acc_list:
        z.append(str(i).count('9'))
    #normalize the list
    normalized_z = []
    for i in z:
        delta = max(z) - min(z)
        normalized_z.append((i - min(z))/delta)
    return normalized_z
```

```python
normalized_ar = get_accuracy_list(100)
plt.xlabel("Iterations(n)")
plt.ylabel("Accuracy(log)")
plt.plot(normalized_ar)
```
